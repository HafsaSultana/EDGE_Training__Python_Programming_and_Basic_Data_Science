```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = {'model':[1,5,6,7]}
df = pd.DataFrame(data, columns=['model'], index=['I1','I2','I3','I4'])
df
```

```python
x = df['model']
plt.figure()
plt.plot(x, '--')
```

```python
data = {'Model1': [86,50,44,13,45,50],'Model2': [77,98,34,15,22,19], 'Model3': [56,23,44,12,32,33]}
df = pd.DataFrame (data, columns = ['Model1','Model2','Model3'], index=['I1', 'I2','I3','I4','I5','I6'])
df
```

```python
x = df['Model1']
y = df['Model2']

plt.figure()

plt.plot(x, color='red', label='Model 1', linewidth= 2, linestyle='--')
plt.plot(y, color='g', label='Model 1', linewidth= 2, linestyle='-')

plt.legend()
```

```python
x = df['Model1']
y = df['Model2']

plt.figure()
plt.plot(x,y, color='g', label='Model1 vs Model2', linewidth=2, linestyle= 'solid')
plt.title("Marge x and y line")

plt.xlim(-10, 100)
plt.ylim(10, 100)

plt.legend()
```

```python
plt.figure()
plt.scatter(x,y)
```

```python
plt.plot(x,y, '-p', color='blue', markersize=15,
         linewidth=4,
         markerfacecolor='red',
         markeredgecolor='gray',
          markeredgewidth=2)
```
