# Using explode

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Tomatoes", "Mangoes", "Oranges", "Apples"]
myexplode = [0.2, 0, 0.1, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

# Barchart with text

```python
import pandas as pd

# list of name, degree, score
nme = ["aparna", "pankaj", "sudhir", "Geeku"]
deg = ["MBA", "BCA", "M.Tech", "MBA"]
scr = [90, 40, 80, 98]

# dictionary of lists
dict = {'name': nme, 'degree': deg, 'score': scr}

df = pd.DataFrame(dict)

print(df)
```

```python
# x  = CSV_data['Product'].tolist()
# y  = CSV_data['Price'].tolist()

x= df['name']
y=df['score']

plt.bar(x,y)

# Change the variable name from 'dict' to something else, like 'my_style'
# my_style = dict(size=10, color='red')

for i,j in zip(x,y):
  plt.text(i, j+1, "({},{})".format(i,j) , ha='center', size=10, color='red')   #**my_style
```
