```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

data = {'Model1': [86,23,32,13,45,12],'Model2': [96,98,34,12,22,19], 'Model3': [56,23,44,12,32,33]}
df = pd.DataFrame (data, columns = ['Model1','Model2','Model3'], index=['I1', 'I2','I3','I4','I5','I6'])
df
```

## Line Plot

```python
x = df['Model1']
plt.figure()
plt.plot(x, '-')
```

```python
x = df['Model1']
plt.figure()
plt.plot(x, '-', color='red', label='Model 1 Plot', linewidth=2, linestyle='--')
plt.legend()
```

```python
x = df['Model1']
y = df['Model2']
plt.figure()
plt.plot(x, color='red', label='Model 1', linewidth=2, linestyle='--')
plt.plot(y, color='g', label='Model 2', linewidth=3, linestyle='dotted')

plt.legend()
```

## Plot against each-other

```python
x = df['Model1']
y = df['Model2']
plt.figure()
plt.plot(x, y, color='red', label='Model 1', linewidth=2, linestyle='solid')


plt.legend()
```

```python
x = df['Model1']
y = df['Model2']
plt.figure()
plt.plot(x, y, color='b', label='Model 1 vs Model 2', linewidth=2, linestyle='solid')
plt.title("Trying Line Plot")

for i_x, i_y in zip(x, y):
    plt.text(i_x, i_y, '({}, {})'.format(i_x, i_y))

plt.xlabel("Model1-axis")
plt.ylabel("Model2-axis")

plt.legend()
```

## Axis Size

```python
x = df['Model1']
y = df['Model2']
plt.figure()
plt.plot(x, y, color='red', label='Model 1', linewidth=2, linestyle='solid')

plt.xlim(-10, 100)
plt.ylim(10, 100)


plt.legend()
```

```python
x = df['Model1']
y = df['Model2']
plt.figure()

plt.plot(x, y, color='blue', label='Model 2', linewidth=2, linestyle='solid')
plt.axis('tight')


plt.legend()
```

```python
x = df['Model1']
y = df['Model2']
plt.figure()

plt.plot(x, y, color='blue', label='Model 2', linewidth=2, linestyle='solid')
plt.axis('equal')


plt.legend()
```
