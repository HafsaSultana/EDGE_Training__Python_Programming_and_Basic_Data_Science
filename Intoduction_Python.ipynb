# Matplotlib

## Line

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.array([3,8,2,10,5])

plt.plot(y, linestyle= '-.',  color='#4c0', linewidth=5)
plt.show()
```

```python
y = np.array([3,44,2,77,10,5])

plt.plot(y, linestyle= '-',  color='#4c0', linewidth=5)
plt.show()
```

```python
y = np.array([3,334,2,77,10,5])

plt.plot(y, linestyle= '-.',  color='#4c1', linewidth=10)
plt.show()
```

```python
y1 = np.array([3,8,2,10])
y2 = np.array([6,4,2,1])

plt.plot(y1, linestyle= '-.',  color='#4c0', linewidth=5)
plt.plot(y2, linestyle= '-',  color='r', linewidth=2)
plt.show()
```

```python
y1 = np.array([3,8,2,10])
y2 = np.array([6,4,2,1])
y3 = np.array([8,4,3,6])

plt.plot(y1, linestyle= '-.',  color='#4c0', linewidth=5)
plt.plot(y2, linestyle= '-',  color='r', linewidth=2)
plt.plot(y3, linestyle= '-',  color='g', linewidth=2)
plt.show()
```

```python
plt.pie(y1)
plt.show()
```

```python
y1 = np.array([3,8,2,10])
plt.hist(y1, density = True)
plt.show()
```
